```python
from Neuron.Neuron import Neuron
from Neuron.NeuronLayer import NeuronLayer
from Neuron.NeuronNetwerk import NeuronNetwerk

import pandas as pd
```

```python
invert_gate = Neuron([-1], 0.5)
and_gate = Neuron([0.5, 0.5], -1)
nor_gate = Neuron([-1, -1, -1], 0)
```

```python
invert_results = pd.DataFrame(
    [
        [0, invert_gate.sigmoid_activation([0]), 1],
        [1, invert_gate.sigmoid_activation([1]), 0],
    ],
    columns= ["X1", "Result", "Target"]
)

invert_results
```

```python
and_results = pd.DataFrame(
    [
        [0, 0, and_gate.sigmoid_activation([0, 0]), 0],
        [1, 0, and_gate.sigmoid_activation([1, 0]), 0],
        [0, 1, and_gate.sigmoid_activation([0, 1]), 0],
        [1, 1, and_gate.sigmoid_activation([1, 1]), 1],
    ],
    columns= ["X1","X2", "Result", "Target"]
)

and_results
```

```python
nor_results = pd.DataFrame(
    [
        [0, 0, 0, nor_gate.sigmoid_activation([0, 0, 0]), 1],
        [1, 0, 0, nor_gate.sigmoid_activation([1, 0, 0]), 0],
        [0, 1, 0, nor_gate.sigmoid_activation([0, 1, 0]), 0],
        [0, 0, 1, nor_gate.sigmoid_activation([0, 0, 1]), 0],
        [1, 1, 0, nor_gate.sigmoid_activation([1, 1, 0]), 0],
        [1, 0, 1, nor_gate.sigmoid_activation([1, 0, 1]), 0],
        [0, 1, 1, nor_gate.sigmoid_activation([0, 1, 1]), 0],
        [1, 1, 1, nor_gate.sigmoid_activation([1, 1, 1]), 0],
    ],
    columns= ["X1", "X2", "X3", "Result", "Target"]
)

nor_results
```
